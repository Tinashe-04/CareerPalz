# Converting Excel Files to CSV

```python
import requests
import os
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dol.gov/agencies/eta/foreign-labor/performance'

response = requests.get(url)
response
```

```python
response.text
```

```python
response.content
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
start_of_url = '/sites/dolgov/files/ETA/oflc/pdfs/'
all_downloads = [link["href"] for link in soup.find_all('a') if link.attrs.get('href') and link["href"].startswith(start_of_url)]
```

```python
len(all_downloads)
```

```python
all_downloads[0]
```

```python
%load_ext autotime
```

```python
list_two = [link for link in all_downloads if link.endswith('.xlsx')]
len(list_two)
```

```python
list_two[0]
```

```python
start_of_url = 'https://www.dol.gov'

print(start_of_url + list_two[0])
```

```python
list_two
```

```python
list_two = [start_of_url + link for link in list_two]
list_two
```

```python
for link in list_two:
    response = requests.get(link)
    filename = os.path.basename(link)
    with open(filename, 'wb') as f:
        f.write(response.content)
```
